```python
import pandas as pd
file_path = 'API_SP.POP.TOTL_DS2_en_csv_v2_81108.csv'
data = pd.read_csv(file_path, skiprows = 4)
```

```python

import matplotlib.pyplot as plt

# Filter countries and 2023 population
selected_countries = ['India', 'China', 'United States', 'Indonesia', 'Brazil', 'Pakistan', 'Nigeria']
data_2023 = data[data['Country Name'].isin(selected_countries)][['Country Name', '2023']]
data_sorted = data_2023.sort_values(by='2023', ascending=False)

# Plot
plt.figure(figsize=(10, 6))
bars = plt.bar(data_sorted['Country Name'], data_sorted['2023'], color='mediumseagreen')
plt.title('Categorical vs Continuous: Country vs Population (2023)')
plt.xlabel('Country (Categorical)')
plt.ylabel('Population (Continuous)')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height, f'{int(height/1e6)}M',
             ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```
